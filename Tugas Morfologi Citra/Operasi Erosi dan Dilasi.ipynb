```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the handwritten text image
image = cv2.imread('jambu.jpg', 0)

# Apply Gaussian blur to reduce noise
image_blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Apply adaptive thresholding to create a binary image
_, binary_image = cv2.threshold(image_blurred, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Define kernel for erosion and dilation
kernel = np.ones((3,3), np.uint8)

# Apply erosion to remove small noise or thin lines
eroded_image = cv2.erode(binary_image, kernel, iterations=1)

# Apply dilation to fill gaps in between letters
dilated_image = cv2.dilate(eroded_image, kernel, iterations=1)

# Display the original image, eroded image, and dilated image
plt.figure(figsize=(10,10))

plt.subplot(1, 3, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(eroded_image, cmap='gray')
plt.title('Eroded Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(dilated_image, cmap='gray')
plt.title('Dilated Image')
plt.axis('off')

plt.show()
```
